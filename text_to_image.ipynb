```python
# Colab Flask App
import torch
from flask import Flask, request, send_file
from pyngrok import ngrok
from io import BytesIO
from diffusers import StableDiffusionPipeline
from PIL import Image

# Set up the ngrok authentication token
ngrok.set_auth_token('thankyou')  # Replace with your ngrok token

# Start a ngrok tunnel
public_url = ngrok.connect(5000).public_url
print(f"To access the global link, please click {public_url}")

# Your model configuration
class CFG:
    device = "cuda" if torch.cuda.is_available() else "cpu"
    seed = 42
    generator = torch.Generator(device).manual_seed(seed)
    image_gen_steps = 35
    image_gen_model_id = "stabilityai/stable-diffusion-2"
    image_gen_size = (400, 400)
    image_gen_guidance_scale = 9

# Load the Stable Diffusion model
image_gen_model = StableDiffusionPipeline.from_pretrained(
    CFG.image_gen_model_id,
    torch_dtype=torch.float16 if CFG.device == "cuda" else torch.float32,
    revision="fp16" if CFG.device == "cuda" else "main",
    use_auth_token='your_hugging_face_auth_token'  # Replace with your Hugging Face token
)
image_gen_model = image_gen_model.to(CFG.device)

# Initialize Flask app
app = Flask(__name__)

# Custom image generation function
def generate_image(prompt, model):
    image = model(
        prompt,
        num_inference_steps=CFG.image_gen_steps,
        generator=CFG.generator,
        guidance_scale=CFG.image_gen_guidance_scale
    ).images[0]
    image = image.resize(CFG.image_gen_size)
    return image

@app.route('/generate_image', methods=['POST'])
def generate_image_endpoint():
    text = request.form.get('text')
    image = generate_image(text, image_gen_model)
    img_io = BytesIO()
    image.save(img_io, 'JPEG')
    img_io.seek(0)
    return send_file(img_io, mimetype='image/jpeg')

if __name__ == '__main__':
    app.run(port=5000)  # Run the Flask app
```
